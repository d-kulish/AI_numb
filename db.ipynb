### Product - category - cluster

```python
# Connection categories/cluster codes/shops 
am_app_cluster_codes = pd.read_sql_query('SELECT * FROM am_app_cluster_codes', engine)

print(am_app_cluster_codes.shape) 

am_app_cluster_codes.head()
```

```python
clust_prod_df = pd.merge(am_app_cluster_codes, am_app_upc_to_cat, left_on='categoryid', right_on = 'category_id', how = 'left')

print(clust_prod_df.shape)

clust_prod_df.head()
```
